# Identifying different types of spenders:

```python
#Creating new column for mean_price per customer.

df_ords_prods_merge['mean_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge[['mean_price','user_id']].head(40)
```

```python
#Creating flag for high and low spenders

df_ords_prods_merge.loc[df_ords_prods_merge['mean_price'] < 10, 'spender_flag'] = 'Low spender'
df_ords_prods_merge.loc[df_ords_prods_merge['mean_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
df_ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

### There are 31,770,614 low spenders and 634,245 high spenders.

# Identifying frequency of customers:

```python
#Creating new column for median frequency per customer.

df_ords_prods_merge['median_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Creating flag for frequency of customers

df_ords_prods_merge.loc[df_ords_prods_merge['median_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['median_frequency'] > 10) & (df_ords_prods_merge['median_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['median_frequency'] < 10, 'frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge['frequency_flag'].value_counts()
```
