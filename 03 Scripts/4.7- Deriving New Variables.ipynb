## Importing libraries

```python
import pandas as pd
import numpy as np
import os
```

```python
#Importing orders/products merged df

df_ords_prods_merged = pd.read_pickle(r'/Users/dorinasalam/Desktop/08-2023 Instacart Basket Analysis/02 Data/Prepared Data/orders_products_merged.pkl')
```

```python
#Only working with a portion of df (1,000,000 rows) 

df = df_ords_prods_merged[:1000000]
```

```python
#Define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Applying the function to the 'price_range' column

df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Determining the count of values in 'price_range' column

df['price_range'].value_counts(dropna = False)
```

```python
#Determining the max value in the 'prices' column
df['prices'].max()
```

```python
#Using the loc() function to create a new column and filter/label the price ranges.

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Determining the count of values in each category.

df['price_range_loc'].value_counts()
```

```python
#Filtering/labeling whole df with the same categories

df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

# Beginning task 4.7: Deriving New Variables

```python
#Assigning a label to the day of the week using a for-loop (adding 2 more labels: very busy & slightly busy)

result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 1:
    result.append("Very busy")
  elif value ==3: 
    result.append("Slightly busy")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#Assigning result to the new column

df_ords_prods_merged['busiest_day'] = result
```

```python
#Checking the count of the values in each category

df_ords_prods_merged['busiest_day'].value_counts()
```

## The frequency follows the same trend as before with the regularly busy days having more frequent orders than the other days of the week. 

## The days with the most sales occur half as frquently as the regularly busy days.

```python
#Determining the busiest hour of the day to create 3 time frames of "Most orders", "Average orders", "Fewest orders"

df_ords_prods_merged['order_hour_of_day'].value_counts()
```

```python
#Assigning the time frames a label

result = []

for value in df_ords_prods_merged['order_hour_of_day']:

  if value in [10,11,14,15,13,12,16,9]:
    result.append('Most orders')
  elif value in [17,8,18,19,20,7,21,22]:
    result.append('Average orders')
  elif value in [23,6,0,1,5,2,4,3]:
    result.append('Fewest orders')
```

```python
#Creating new column and applying it to the original df. 

df_ords_prods_merged['busiest_period_of_day'] = result
```
